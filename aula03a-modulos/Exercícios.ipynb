```python
# Exercício 2 discernir se um número é impar ou par
import random
num = random.randint(0,1024)
print("número:", num)

if num % 2 == 1:
    print('Número impar')
else:
    print('Número par')
```

```python
# Exercício 3 input 2 números e qual o maior ou se são iguais
num1 = input("número 1:")
num2 = input("número 2:")

if num1 > num2:
    print("O primeiro número é maior")
elif num1 < num2:
    print("O segundo número é maior")
else:
    print("Os número são iguais")

```

```python
# Exercício 3 mais legal
import random

num1 = random.randint(0,1024)
num2 = random.randint(0,1024)
tentativas = 0
while num1 != num2:
    tentativas += 1

    num1 = random.randint(0,1024)
    num2 = random.randint(0,1024)

    print('número 1:', num1)
    print('número 2:', num2)
    
    if num1 > num2:
        print("O primeiro é maior")
    elif num1 < num2:
        print("O segundo é maior")
    else:
        print("Nossa, qual a chance?")
print('Tentativas:',tentativas)
```
