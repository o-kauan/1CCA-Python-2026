```python
# Exercício 2 discernir se um número é impar ou par
import random
num = random.randint(0,1024)
print("número:", num)

if num % 2 == 1:
    print('Número impar')
else:
    print('Número par')
```

```python
# Exercício 3 mais legal
import random

num1 = random.randint(0,1024)
num2 = random.randint(0,1024)
tentativas = 0
while num1 != num2:
    tentativas += 1

    num1 = random.randint(0,1024)
    num2 = random.randint(0,1024)

    print('número 1:', num1)
    print('número 2:', num2)
    
    if num1 > num2:
        print("O primeiro é maior")
    elif num1 < num2:
        print("O segundo é maior")
    else:
        print("Nossa, qual a chance?")
print('Tentativas:',tentativas)
```

```python
# Exercício 4 ler 4 notas, fazer a média e dizer se passou, está de rec ou reprovou
import random

list = []
for i in range (1,5):
    nota = random.randint(0,10)
    list.append(nota)
    print(f'nota{i}: {nota}')

med = sum(list) / len(list)
print('Média:', med)   

if med >= 7:
    print("PASSOU")
elif med >= 5:
    print("RECUPERAÇÃO KKKKKK")
else:
    print('SE LASCOU')
```

```python
# Exercíco 5 ler 2 int, discernir se são multiplos
import random

num1 = random.randint(0,1024)
num2 = random.randint(0,1024)

print('número 1:', num1)
print('número 2:', num2)

if num1 % num2 == 0 or num2 % num1 == 0:
    print('múltiplos')
else:
    print('não múltiplos')
```
