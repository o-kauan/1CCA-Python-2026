```python
# Exercício 4 ler 4 notas, fazer a média e dizer se passou, está de rec ou reprovou
import random

list = []
for i in range (1,5):
    nota = random.randint(0,10)
    list.append(nota)
    print(f'nota{i}: {nota}')

med = sum(list) / len(list)
print('Média:', med)   

if med >= 7:
    print("PASSOU")
elif med >= 5:
    print("RECUPERAÇÃO KKKKKK")
else:
    print('SE LASCOU')
```

```python
# Exercício 8 recebe um salário e reajusta: até 280 (incluido) + 20%, entre 280 e 700 + 15%, entre 700 e 1500 + 10% e 1500 pra cima + 5%
# apresentar salário antes do reajuste, percentual do aumento aplicado, valor do aumento e novo salaário com aumento
import random

def br(valor):
    return f"{valor:,.2f}".replace(',', '#').replace('.', ',').replace('#', '.')

sal = random.randint(200, 2000)
indce = 0

if sal <= 280:
    indce = 0.2
elif sal <= 700:
    indce = 0.15
elif sal <= 1500:
    indce = 0.1
else:
    indce = 0.05

bon = indce * sal
sal_novo = sal + bon

# Formatação para moeda brasileira
sal_br = br(sal)
bon_br = br(bon)
sal_novo_br = br(sal_novo)

print(f"O salário bruto é: R${sal_br}, o percentual de aumento é: {indce:.0%}\nO valor do aumento é: R${bon_br} e o novo salário é de R${sal_novo_br}.")
```

```python
# Exercício 9 recebe: estado (1 a 5), peso em T, cód (10 a 40) cal: peso p/ quilos, preço da carga, vlr imposto (preço e origem) e valr total (carga + imposto))
import random 

estado = random.randint(1,5)
peso_emT = round(random.uniform(1, 70),4)
cod = random.randint(10,40)

peso = peso_emT * 1000

# cal preço da garga considerando peso e código
preco_kg = 0
if cod <= 20:
    preco_kg = 100
elif cod <= 30:
    preco_kg = 250
else:
    preco_kg = 340
preco_garga = preco_kg * peso

# cal imposto
imposto = 0
if estado == 1:
    imposto = 0.35
elif estado == 2:
    imposto = 0.25
elif estado == 3:
    imposto = 0.15
elif estado == 4:
    imposto = 0.05
else:
    imposto = 0

imposto_carga = preco_garga * imposto

valor_total = preco_garga + imposto_carga

# formatação
def br(valor):
    return f"R${valor:,.2f}".replace(',', '#').replace('.', ',').replace('#', '.')

if imposto_carga == 0:
    imposto_carga = "Isento"
else:
    imposto_carga = br(imposto_carga)

peso = f"{peso:,.2f}".replace(',', '#').replace('.', ',').replace('#', '.')
peso_emT = f"{peso_emT:,.4f}".replace(',', '#').replace('.', ',').replace('#', '.')

print(f"Estado: {estado}\nPeso: {peso_emT}T ({peso} kg)\nCódigo: {cod} (R${(preco_kg)} por kg)\nPreço da carga: {br(preco_garga)} \nImposto: {imposto_carga} ({imposto:.0%})\nValor total: {br(valor_total)}")
```

```python
# Exercício 10:
lado_1 = float(input("Digite o valor do lado 1: "))
lado_2 = float(input("Digite o valor do lado 2: "))
lado_3 = float(input("Digite o valor do lado 3: "))

lados = sorted([lado_1, lado_2, lado_3], reverse=True)
A, B, C = lados

if A >= B + C:
    print("NAO FORMA TRIANGULO")
else:
    if A * A == B * B + C * C:
        print("TRIANGULO RETANGULO")
    elif A * A > B * B + C * C:
        print("TRIANGULO OBTUSANGULO")
    else:
        print("TRIANGULO ACUTANGULO")

    if A == B == C:
        print("TRIANGULO EQUILATERO")
    elif A == B or B == C or A == C:
        print("TRIANGULO ISOSCELES")
```
